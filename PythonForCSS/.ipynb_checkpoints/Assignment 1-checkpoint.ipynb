_2. Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order._

```python
# Write your solution here
import pandas as pd
import numpy as np


# Define an array and a target value
arr1 = np.array([1,5,6,6,7,134,200, 250, 256, 300])
my_target = 250


# The target finder/inserter function

def target_val(target, my_array):   
    try:
        target = float(target)
    except:
        print("Please follow the instructions.")         
        
    if target in my_array:
        #for finding only the first occurence.
        row_index = np.where(my_array == target)[0][0]
        print(f"The target is found at index {row_index}.")
    else:
        expected_row_index = np.where(my_array > target)[0][0]
        my_array = np.insert(my_array, expected_row_index, target)
        print(f"Target is not found and added to its expected place at index {expected_row_index}.")



target_val(my_target, arr1)
```

_3. Given two binary strings a and b, return their sum as a binary string._

For instance:

- Input: a = "11", b = "1"

- Output: "100"


```python
# Write your solution here
import numpy as np


# I designed the workflow as follows:


# To calculate two values as binary string numbers.

# First make the numbers lists.
# Check if the values are in binary digits

# Then iterate backwards to compute digit index
# Because index number is in reverse order with digit power
# Get summation by multiplying 2power to the index

# Append starting from the highest digit
# Results 


def binary_sum(a,b):
    print("You've entered", a, "and", b, "values.")    
    a = str(a)
    b = str(b)
    # If you entered wrong values such as numbers not 1 or 0
    # or string values, function raises error.
    
    try: 
        # Setting up variables
        aL = [int(i) for i in a]
        bL = [int(i) for i in b]          
        aL_sum = 0
        bL_sum = 0
        binary_sum = 0
        aL.reverse()
        bL.reverse()    
        sumL = []
        sumL_final = ""
        
        #Calculating summation for binary values
        for i in range(len(aL)):
            if aL[i] == 0 or aL[i] == 1:
                aL_sum = aL_sum + aL[i]*2**i
            else:
                aL_sum = None                 
        for i in range(len(bL)):
            if bL[i] == 0 or bL[i] == 1:
                bL_sum = bL_sum + bL[i]*2**i
            else:
                bL_sum = None   
        
        # Find the final sum
        the_sum = aL_sum + bL_sum
        
        
        # Change final sum to binary sum
        
        # First assess digit count
# I blundered range, usin len*len yields 
#wrong result for single digits...
        for i in range(len(bL)+len(aL)):
            if the_sum - 2**i >= 0:
                our_digit = int(i)
            else:
                break
        # Then sum up to digits    
        for i in range(our_digit+2):
            if our_digit == -1:
                break
            elif the_sum - 2**our_digit >= 0:
                sumL.append("1")
                the_sum = the_sum - 2**our_digit
                our_digit = our_digit - 1
            elif the_sum - 2**our_digit < 0:
                sumL.append("0")
                our_digit = our_digit - 1  
        
        # Changing list to a string
        for i in sumL:
            sumL_final = sumL_final + i
        
        print("Final sum is:", sumL_final)
            
    except:
        print("Enter correct value")
 

        
     



binary_sum(111,111)





```

_4. Given a non-empty array of integers numbers, every element appears twice except for one. Find that single one._

```python
# Write your solution here
import random
import numpy as np


# I created a non_empty array of integer numbers, 
# every element appears twice, except the value 99

a = np.arange(100)
a = np.append(a,a[:-1])

#print(a)

# I created another array of unique values
# and counts, of the array a

b, counts = np.unique(a, return_counts = True)
#print(b)
#print(counts)

# then I found which number has a single occurrence

print("The number that has unique occurrence:",b[counts == 1])
```

_5. The DNA sequence is composed of a series of nucleotides abbreviated as 'A', 'C', 'G', and 'T'._

For example, "ACGAATTCCG" is a DNA sequence.
When studying DNA, it is useful to identify repeated sequences within the DNA.

Given a string s that represents a DNA sequence, return all the 10-letter-long sequences (substrings) that occur more than once in a DNA molecule. You may return the answer in any order.

For instance:

- Input: s = "AAAAACCCCCAAAAACCCCCCAAAAAGGGTTT"

- Output: ["AAAAACCCCC","CCCCCAAAAA"]


```python
# Write your solution here


# Our target example
dna_string1 = "AAAAACCCCCAAAAACCCCCCAAAAAGGGTTT"

# My test example
dna_string2 = "AAATTTTTAAAAAAGGGGGCCCCCAAAAACACCTTTTTCCCCCAGGAAGGAAGGCCCCCAAAAAGTTT"


# To solve the problem, I created a sample list
# to search if the string has matching elements

dna_samp = ["AAAAACCCCC", "CCCCCAAAAA", "AAAAAGGGGG", "GGGGGAAAAA", "AAAATTTTT","TTTTTAAAAA", "CCCCCGGGGG","GGGGGCCCCC", "CCCCCTTTTT","TTTTTCCCCC", "GGGGGTTTTT","TTTTTGGGGG"]

# I created an empty list to capture matching elements.
dna_capture = []



# Then iterated through the search example

for i in range(len(dna_string1)):
    try:
        if dna_string1[i:i+10] in dna_samp:
            #print(dna_string1[i:i+5])
            dna_capture.append(dna_string1[i:i+10])
        else:
            continue
    except:
        continue

print(dna_capture, ": Question example")

dna_capture.clear()

for i in range(len(dna_string2)):
    try:
        if dna_string2[i:i+10] in dna_samp:
            #print(dna_string1[i:i+5])
            dna_capture.append(dna_string2[i:i+10])
        else:
            continue
    except:
        continue

# The answer
print(dna_capture, ": My test example")
    
```
