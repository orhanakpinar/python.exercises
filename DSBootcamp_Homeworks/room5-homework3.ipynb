female_male_ratio ortalaması nedir? -6

```python
df.dtypes

df1 = df

df1.dropna(subset=['female_male_ratio'], inplace=True)

split_ratios = df1['female_male_ratio'].str.split('[:|-]', expand=True)

ratio_values = split_ratios.apply(pd.to_numeric)

ratio_values = ratio_values[(ratio_values[0] != 0) & (ratio_values[1] != 0)]

mean_ratio = (ratio_values[0] / ratio_values[1] ).mean()

print("Female_male_ratio ortalaması:", mean_ratio)


```

3- world_rank 1-100 arasında olan üniversitelerin ortalama international puanı nedir? -3

```python
ilk_100=df[df['world_rank']<'100']['university_name'].unique()
toplam=0
for i in ilk_100:
    deneme=df[df['university_name']==i]['international'].value_counts().reset_index()
    deneme["international"]=pd.to_numeric(deneme['international'])
    toplam=toplam + deneme["international"].mean()
    toplam
print("1-100 Ünilerin ortalama international puanı:",toplam)
```

4- Görselleştirme: En yüksek 10 teaching puanına sahip üniversiteleri ve bu puanları gösteren bir bar grafiği oluşturunuz.


```python
#en yüksek 10 teaching i aradığımız için nlargest fonksiyonu işimizi görecektir
barplot=df.nlargest(10, 'teaching')[['university_name', 'teaching']]
barplot.plot(kind='barh', x='university_name', y='teaching', figsize=(10, 5)).invert_yaxis()
plt.title('Top 10 University teaching points')
```

9- world_rank değeri 50'nin altında olan üniversitelerin ortalama income değeri nedir?



```python
df['income'] = df['income'].replace('-', np.nan)
world_ranks=df[df['world_rank']<"50"][['university_name','income']]
world_rank_mean=pd.to_numeric(world_ranks['income']).mean()
print("50'nin altında olan üniversitelerin ortalama income değeri: ",world_rank_mean)
```
